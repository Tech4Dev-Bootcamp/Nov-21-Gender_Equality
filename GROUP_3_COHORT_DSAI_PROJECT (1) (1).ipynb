# TITLE: GENDER GAP IN EDUCATION

## INTRODUCTION

Gender equality is when people of all genders have equal rights, responsibilities and opportunities. Everyone is affected by gender inequality - women, men, trans and gender diverse people, children and families. It impacts people of all ages and backgrounds.

Gender equality in education simply refers to equal and fair right to education to all gender, women, men, trans and gender diverse people

In our project we are analysing the *Gap of gender equality in education*.
Is fair education being given to all gender equally? What is the ratio of education being given between boys and girls? Which countries have low ratios and which countries have high ratios? Do we still have a lot of work to do before SDG Goals 4 & 5 are achieved?

## DATASET
The Dataset used for this project was gotten from Gapminder [here](https://www.gapminder.org/data/). The dataset is titled 'Gender Ratio of Mean Years in School (Women % Men, 25 - 34 years)'. The dataset gives the mean ratio of the number of years spent in school by Males and Females within the age range of 25 - 34 among different countries. The data is compiled from the year 1970 to 2015.

## INSIGHTS/FINDINGS

1. In the 188 countries listed, the five countries in the lead in terms of Gender ratio are:Lesotho, Qatar, Botswana, Antigua and Barbuda, and Moldova.	Surprisingly, two African countries - Lesotho and Bostwana are in the lead. 
2. Lesotho is the highest in terms of female education in the world. Although on the surface this is a great achievement, a deeper look into the cause might prove otherwise. According to a blog post in 2018, [here](https://borgenproject.org/female-education-in-lesotho/) this ratio is as a result of emigration to South Africa. Due to high unemployment and poverty rates, majority of the men in Lesotho emigrated to South Africa. Therefore in Lesotho women outnumber men at a ratio of four-to-one. Most males left behind withdraw from school to rear cattle and take care of their family or withdraw from school to emigrate to South Africa. This results a registration number of 1.6 females for every male. Lesotho began with a gender ratio of 129.0 in 1970. Then it increased one point value to 130.0 and maintained this until 1983. From then onwards, the value declined steadily until 126.0 in 2015. This means that more girls are either not being enrolled in schools as the years go by or girls are steadily withdrawing from school.
3. Botswana leads in second place after Lesotho in terms of female education in the world. The government has made access and equality of opportunity for education an explicit policy (Republic of Botswana, 1977; RNPE, 1994). Several studies have shown that girls tend to have a positive attitude towards school and conform to school expectations compared to boys (Darling and Glendenning, 1996; Dunne and Leach, 2005). Furthermore, some studies have shown differential treatment of learners in the classroom, based on gender. For instance, Foster et al. (1996) and Woolfolk (2004) state that boys’ misbehaviour result in teachers giving them more attention compared to girls. The unequal treatment affects student’s motivation and self-esteem with boys dominating in class discussions whilst girls underestimate their academic ability. Generally, most studies have shown that girls are disadvantaged as they are asked less questions and teachers have lower expectation of them compared to boys. Girls also suffer from sexual harassment from boys and teachers (UNICEF, 2000, Kamwendo, 2010). Draft report on the development of education ‘inclusive education: the way of the future.
http://www.docstoc.com/docs/30987873/draft-report-republic. Retrieved 13/1/2011.

4. In Southern African countries, South Africa does not meet expectations in terms of gender ratios. This is because of racial disparities in the educational system and wealth. Racial inequality still play a major role and black girls are at a disadvantage in terms of education. In a country made up of a large population of blacks, this poses a major problem. Other factors such as gender based violence and patriachy seeks to ensure females do not attend school.
5. The countries with the lowest Gender ratios are: Guinea,South Sudan, Chad, Yemen and  Afghanistan with the ratios decreasing respectively. In these countries some of the major reasons why girls do not attend school include religious and cultural beliefs, poverty and safety concerns. In places like Afghanistan and Yemen, mixed classes are ruled against by either the Taliban or religious leaders. Since there is no provision for separate classes, most families forgo educating their daughters. In the case of Chad, cultural beliefs and early marriage prevent girls from going to school. In Guinea, these cultural beliefs play a dominant role as it is believed that the girls are more useful at home, taking care of the family/doing house chores while the boys go to school. These same factors affect girls in South Sudan as Gender bias continue to play a predominant role.
6. In countries like Mauritania and Seychelles, which take the lead in terms of education in their respective region. Their government has seen the value of educating thr female child and are constantly seeeking partnerships with major world organization to bridge the gap and empower the female child.


## CONCLUSION

There is still a lot of work to be done in Africa before we can achieve SDG goals 4 & 5 by 2030. The government of these countries need to carry out local sensitization to help educate locala families on the importance of educating the girl child. More schools with good access roads/proper network and ammenities should be built to cater to the ever-growing population. In places where religious beliefs may play a part, the government and local community can consider building separate classrooms until a better solution comes up so girls are not at a disadvantage.
